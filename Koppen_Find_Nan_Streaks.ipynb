```python
import numpy as np
import pandas as pd

def find_nan_streaks(df, value_col='기온(°C)'):
    df = df.copy()
    if '일시' not in df.columns or value_col not in df.columns:
        return []

    df['일자'] = pd.to_datetime(df['일시']).dt.date
    daily = df.groupby('일자')[value_col].mean()

    is_nan = daily.isna().astype(int)

    streaks = []
    count = 0
    for val in is_nan:
        if val == 1:
            count += 1
        else:
            if count > 0:
                streaks.append(count)
                count = 0
    if count > 0:
        streaks.append(count)
    
    return streaks

longest_streaks_per_file = []

for i in range(len(temp)):
    for j in range(len(temp[i])):
        df = temp[i][j]
        streaks = find_nan_streaks(df, value_col='기온(°C)')
        longest = max(streaks) if streaks else 0
        longest_streaks_per_file.append({
            '연도': 1980 + i * 5 + j,
            '최장 결측일 연속': longest
        })

streaks_df = pd.DataFrame(longest_streaks_per_file)
```
